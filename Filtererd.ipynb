```python
import csv
import pandas as pd
df= pd.read_csv("RP_webscrap.csv",engine='python')
n=len(df['Paper'])
#print(n)
keyword=str()
paper=str()
url=str()
names=str()
abstract=str()
data=[[keyword,paper,url,names,abstract]]
i=0
while(i<n):
    j=i+1
    
    while(j<n):
            if (df.Paper[i]==df.Paper[j]) or (df.Names[i]==df.Names[j]):
               
                i=i+1
                break
            elif (df.Paper[i]!=df.Paper[j]) and (df.Names[i]!=df.Names[j]):
                
                if j+1==n :   
                    
                    keyword=df.Keyword[i]
                    paper=df.Paper[i]
                    url=df.URL[i]
                    names=df.Names[i]
                    abstract=df.Abstract[i]
                    data.insert(i,[keyword,paper,url,names,abstract])
                    new_df=pd.DataFrame(data,columns = ['keyword','paper','url','names','abstract'])
  
                    new_df.to_csv("Filtered.csv")
                    
    
                    break
                else:
                    j=j+1
                    continue
                    

    i=i+1               
                        
    



                
```
